```python
def load_data(city, month, day):
    
    df = pd.read_csv(CITY_DATA[city])
    # convet start_time to datetime 
    df['start_time'] = pd.to_datetime(df['start_time'])
    #extract month and day from start time to create new columns
    df['month'] = df['start_time'].dt.month 
    df['day of week'] = df['start_time'].dt.day_name()

    #filter by month 

    if month != 'all' :
        #use index for months to create corresponding int 
        months = ['january', 'february','march','april','may','june']
        month = months.index(month) + 1 
    
        #filter for month to craete new df 
        df=df[df['month'] == month]
     
        
    #filter by day 
    if day != 'all' :
        #use index for day to create new df 
        df=df[df['day_of_week']== day.title()]
    
    return df
```

```python
def time_stats(df):
    """Displays statistics on the most frequent times of travel."""

    print('\nCalculating The Most Frequent Times of Travel...\n')
    start_time = time.time()

    # TO DO: display the most common month
    common_month = df['month'].mode()[0]
    print('most common month :', common_month)
    
    

    # TO DO: display the most common day of week
    common_day  = df['day of week'].mode()[0]
    print['most common day :',common_day]

    # TO DO: display the most common start hour
    #extract hour from start_time to get a hour column 
    df['hour'] = df['start_time'].dt.hour 
    common_hour = df['hour'].mode()[0]
    print('most column start hour :', common_hour)


    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
    
```
